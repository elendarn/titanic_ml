## Read and Explore Data

# STEP1: Start with the Pclass column only (because it is numerical and complete).

```python
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

blackbox=pd.read_csv("./data/train.csv")
```

```python
# blackbox.info()
```

```python
# blackbox.head(5)
```

```python
df1 = blackbox[['Pclass','Survived']]
df1 
```

```python
blackbox['Survived'].unique()
```

```python
X = df1[['Pclass']]
y = df1['Survived']
```

```python
X.shape, y.shape
```

## Test split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1000)
```

```python
m=LogisticRegression()
m
```

```python
m.fit(X_train, y_train)
```

```python
m.score(X_train, y_train)
```

```python
m.score(X_test, y_test)
```

# STEP2: Print the coefficients calculated by the model.

```python
m.coef_, m.intercept_
```

# STEP3: Calculate the probabilities for your data points belonging to the positive class.

```python
m.predict_proba(X_test)
```
